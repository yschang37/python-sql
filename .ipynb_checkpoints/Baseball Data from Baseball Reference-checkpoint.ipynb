# This python file is to show how to use beaufisoup library to scrape the data      from 
https://www.baseball-reference.com/players/t/troutmi01.shtml


I write the code to get the Mike Trout's Standard Batting.


```python
import requests
from bs4 import BeautifulSoup
import sqlite3
import pandas as pd
import numpy as np

# this function is to output the dataframe
def StandardBatting():
    data = []
    #get the url
    url = 'https://www.baseball-reference.com/players/t/troutmi01.shtml'
    #use requests to get the data from url
    res = requests.get(url)
    #use html.paser to know the detail
    soup = BeautifulSoup(res.text, 'html.parser')
    #this col is to rename the columns
    col = []
    
    #get the detail about the columns
    for sta in soup.find_all('table', attrs={'class':'row_summable sortable stats_table'}):

        for sta_detail in sta.find_all('thead'):

            for sta_detail_2 in sta_detail.find_all('tr'):
                for sta_detail_3 in sta_detail.find_all('th'):
                    col.append(sta_detail_3.text)

    # get the data from table
    cc = pd.DataFrame([])
    df = pd.DataFrame([])
    for sta in soup.find_all('table', attrs={'class':'row_summable sortable stats_table'}):
        for sta_detail in sta.find_all('tbody'):
            for sta_detail_2 in sta_detail.find_all('tr'):
                data = []
                data.append(sta_detail_2.find('th').text)
                for sta_detail_3 in sta_detail_2.find_all('td'):
                    data.append(sta_detail_3.text)
                    if len(data) >= 30:
                        df = df.append([data],ignore_index=True)
    df.rename(index=str, columns={int(i):col[i] for i in range(30)})
    df.columns = col
    print(df)
    return df
```
